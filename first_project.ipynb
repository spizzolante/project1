```python
# function to make all column names to lowercase
def update_columns_to_lower(df):
    df.columns = df.columns.str.lower().str.strip()
    return df


# function to convert all floats as integers
def float_to_int(df):
    for column in df.columns:
        if df[column].dtype == 'float64':
            df[column] = df[column].astype(int)
    return df
```

```python
cleaned_df = wcm_df.dropna(thresh=3)
```

```python
cleaned_df.shape
```

```python
cleaned_df.tail()
```

```python
attendance_mean = wcm_df['Attendance'].mean()
print(attendance_mean)
```

```python
cleaned_df['Attendance'] = cleaned_df['Attendance'].fillna(attendance_mean)
```

```python
cleaned_df = float_to_int(cleaned_df)
```

```python
cleaned_df = update_columns_to_lower(cleaned_df)
```

```python
cleaned_df.dtypes
```

```python
spain_home_matches = cleaned_df[cleaned_df['home team name'] == 'Spain']
spain_away_matches = cleaned_df[cleaned_df['away team name'] == 'Spain']
spain_matches = pd.concat([spain_home_matches, spain_away_matches])
```

```python
spain_df = cleaned_df[(cleaned_df['home team name'] == 'Spain') | (cleaned_df['away team name'] == 'Spain')]
```

```python
spain_df['team'] = spain_df['home team name'].where(spain_df['home team name'] == 'Spain', spain_df['away team name'])
```

```python
spain_df['opponent'] = spain_df['home team name'].where(spain_df['home team name'] != 'Spain', spain_df['away team name'])
```

```python
spain_df['is home team'] = spain_df['home team name'] == 'Spain'
```
